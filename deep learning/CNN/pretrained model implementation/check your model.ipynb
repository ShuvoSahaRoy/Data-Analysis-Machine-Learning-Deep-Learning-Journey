```python
import os
import numpy as np

# Keras
from tensorflow.keras.applications.imagenet_utils import preprocess_input
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image

# Model saved with Keras model.save()
MODEL_PATH = r'E:\codes\python\Web_development\my projects\Computer Vision with Flask\models\Alzheimer.h5'

# Load your trained model
model = load_model(MODEL_PATH)

directory = r'D:\image classification\Alzheimer\test\Very_Mild_Demented'
files = os.listdir(directory)
len(files)
```

```python
# ['Mild_Demented', 'Moderate_Demented', 'Non_Demented', 'Very_Mild_Demented']
pre = list()
Mild_Demented = 0
Moderate_Demented = 0
Non_Demented = 0
Very_Mild_Demented = 0

for i, file in enumerate(files):
    img_path = os.path.join(directory, file)
    img = image.load_img(img_path, target_size=(224, 224))
    
    # Preprocessing the image
    x = image.img_to_array(img)
    # x = np.true_divide(x, 255)
    x = np.expand_dims(x, axis=0)
    
    # Be careful how your trained model deals with the input
    # otherwise, it won't make correct prediction!
    # x = preprocess_input(x)
    
    preds = model.predict(x) # 
    pred_class = preds.argmax(axis=-1) 
    pre.append(list(pred_class))
    
    if pre[i][0] == 0:
        result = "Mild_Demented"
        Mild_Demented = Mild_Demented+ 1
    elif pre[i][0] == 1:
        result = "Moderate_Demented"
        Moderate_Demented = Moderate_Demented + 1
    elif pre[i][0] == 2:
        result = "Non_Demented"
        Non_Demented = Non_Demented + 1
    elif pre[i][0] == 3:
        result = "Very_Mild_Demented"
        Very_Mild_Demented = Very_Mild_Demented + 1
print(f"Mild_Demented {Mild_Demented} Moderate_Demented {Moderate_Demented}\
        Non_Demented {Non_Demented} Very_Mild_Demented{Very_Mild_Demented}")
```
