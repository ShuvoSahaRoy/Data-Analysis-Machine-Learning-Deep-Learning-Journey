# Probabilities

```python
from numpy.random import seed, randint

seed(5)

def coin_toss(flips):
  heads = 0
  tails = 0
  for i in range(0,flips):
    if randint(0, 2) == 1:
        heads += 1
    else:
        tails += 1
  print(f"{heads} {tails}")
  return heads/flips

coin_toss(1000)
```

### Let's store the results of coin toss simulations

```python
probabilities = []
for i in range(1,1000):
  results = (i,coin_toss(i))
  probabilities.append(results)
```

```python
# Convert our data to a dataframe
import pandas as pd

df = pd.DataFrame(probabilities)
df.columns = ['Number of Coin Tosses', 'Probability']
df.head(15)
```

```python
# Now let's create our plot
import matplotlib.pyplot as plt
import seaborn as sns

sns.relplot(x="Number of Coin Tosses", y="Probability", size=1, kind="line", data=df,errorbar=None)
plt.axhline(0.5, color = 'Red', label = '0.5')
```
