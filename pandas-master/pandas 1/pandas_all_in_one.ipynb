### 2010-2014 suicide rate between males and females in swizerland

## Male

```python
mask2 = df.year.between(2010,2014)  # data between 1989 to 2000
mask3 = df.sex== 'male'
mask4 = df.country == 'Switzerland'
mask5 = df.age == '15-24 years'
f_df = df.loc[mask2 & mask3 & mask4 & mask5, ['country','year','sex','age', 'suicides_no']]
f_df
```

```python
# f_df.suicides_no.sum()
f_df.sum().suicides_no
```

## Female

```python
mask2 = df.year.between(2010,2014)  # data between 1989 to 2000
mask3 = df.sex== 'female'
mask4 = df.country == 'Switzerland'
mask5 = df.age == '15-24 years'
f_df = df.loc[mask2 & mask3 & mask4 & mask5, ['country','year','sex','age', 'suicides_no']]
f_df
```

```python
# f_df.suicides_no.sum()
f_df.sum().suicides_no
```
