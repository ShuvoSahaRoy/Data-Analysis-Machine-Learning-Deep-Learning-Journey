## Converter

```python
def convert_people_cell(cell):
    if cell=="n.a.":
        return 'Sam Walton'
    return cell

def convert_price_cell(cell):
    if cell=="n.a.":
        return 50
    return cell
    
df = pd.read_csv("stock_data.csv", converters= {
        'people': convert_people_cell,
        'price': convert_price_cell
    })
df
```

### Fillna

```python
new_df = df.fillna(0)
new_df
```

```python
new_df = df.fillna({
    'temperature': 0,
    'windspeed' : 0,
    'event' : 'no event'
})
new_df
```

```python
new_df = df.fillna({
    'temperature': df.temperature.mean(),
    'windspeed' : 0,
    'event' : 'no event'
})
new_df
```

```python
new_df = df.fillna(method = "ffill")  # forward fill
new_df
```

```python
# new_df = df.fillna(method = "bfill")           # backward fill 
# new_df = df.fillna(method = "bfill" , axis = "columns" )    ## we can use  # axis = "columns" or 'rows'
new_df = df.fillna(method = "bfill" , limit = 1 )  # it will copy the result into only one cell
new_df
```

```python
new_df = df.interpolate()    # linear interpollation 
new_df
```

```python
new_df = df.interpolate(method='time')    # linear interpollation 
new_df
```

```python
new_df = df.dropna()
new_df
```

```python
new_df = df.dropna(how="all")
new_df
```

```python
new_df = df.dropna(thresh=1)
new_df
```
